```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
        n_samples    = 500,
        n_features   = 2,
        centers      = 3,
        cluster_std  = 1.2,
        shuffle      = True,
        random_state = 0
       )
```

```python
from sklearn.model_selection import train_test_split

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Ahora puedes utilizar X_train y y_train para entrenar tu modelo y X_test y y_test para evaluar su rendimiento.
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

# Definir el modelo
model = MLPClassifier()

# Definir los hiperparámetros que queremos ajustar y los valores a probar
param_grid = {
    'hidden_layer_sizes': [(2,5), (10,15), (25,25)],
    'activation': ['relu', 'tanh', 'logistic'],
    'alpha': [0.0001, 0.001, 0.01],
}

# Crear el objeto GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5)

# Realizar la búsqueda de hiperparámetros
grid_search.fit(X, y)

# Ver los mejores hiperparámetros encontrados
best_params = grid_search.best_params_
print(f"Mejores hiperparámetros: {best_params}")
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

# Definir el modelo
model = MLPClassifier()

# Definir los hiperparámetros que queremos ajustar y los valores a probar
param_grid = {
    'hidden_layer_sizes': [(10,),(15,),(25,),(2,5), (10,15),(25,25)],
    'activation': ['relu'],
    'alpha': [0.001],
}

# Crear el objeto GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5)

# Realizar la búsqueda de hiperparámetros
grid_search.fit(X, y)

# Ver los mejores hiperparámetros encontrados
best_params = grid_search.best_params_
print(f"Mejores hiperparámetros: {best_params}")
```

#Análisis-Resultados
Según el método GridSearch con la configuración asignada pudimos encontrar que la mejor configuración de hiperparámetros para la RNA es función de activación:'relu', alpha:0.001, número de capas:(25,25) pero al comparar cada split por número de capas vs accuracy pudimos notar que hay combinaciones de capas muy buenas e incluso más simples (sin tantas neuronas e incluso capas) por lo que según nuestro criterio creeríamos que es mejor trabajar con 1 capa de 25 neuronas y la diferencia de porcentaje de accuracy es 0.002.

```python
# Definir el modelo con los mejores hiperparámetros encontrados
best_params = {
    'hidden_layer_sizes': (25,25),  # Por ejemplo, aquí usamos 50 neuronas en la capa oculta
    'activation': 'relu',         # Por ejemplo, aquí usamos la función de activación ReLU
    'alpha': 0.001               # Por ejemplo, aquí usamos alpha=0.001 para regularización
}

model = MLPClassifier(**best_params)

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

# Predecir las etiquetas del conjunto de prueba
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Calcular la exactitud en el conjunto de prueba
accuracy = accuracy_score(y_test, y_pred)

print(f"Exactitud del modelo: {accuracy}")
```
